```python
import urllib
import scipy.io.wavfile
import pydub

# read wav file
rate, audData = scipy.io.wavfile.read("/home/ilkyaz/Downloads/Valse.wav")

# sampling rate is the number of data points sampled per second in the audio file
# 44100 pieces of information per second make up the audio wave

print("Sampling Rate: ", rate)
print("Audio Wave Data: \n", audData)
```

```python
#wav length
audData.shape[0] / rate
```

```python
import numpy as np

# there is separate data for the left and right channels in the audio file
channel1 = audData[:,0] #left
channel2=audData[:,1] #right

#Energy of music
np.sum(channel1.astype(float)**2)
```

```python
import matplotlib.pyplot as plt

# create a time variable in seconds
time = np.arange(0, float(audData.shape[0]), 1) / rate

# plot amplitude (or loudness) over time
plt.figure(1)
plt.subplot(211)
plt.plot(time, channel1, linewidth=0.01, alpha=0.7, color='#ff0080')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.subplot(212)
plt.plot(time, channel2, linewidth=0.01, alpha=0.7, color='#ff0080')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```
